```python
import numpy as np
from skimage.feature import hog
import os
import cv2

def extract_hog_top(images,orientations = 9,pixels_per_cell = (8, 8),cells_per_block=(2,2), eps=1e-7):
        #Load image sequence
        hogtop_features = [ ]
        hognum=0
        video_length = len(images)
        #print(images.shape) shows (1075,128,128)
        for i in range(images.shape[0]-k):
            first_frame = i
            last_frame= i + window_size
            #print(first_frame,last_frame)
            first_img = images[first_frame]
            last_img=images[last_frame]
            #print(first_img.shape)
            #print(last_img.shape)
            first_img = cv2.resize(first_img, (128, 64))
            last_img = cv2.resize(last_img, (128, 64))
            #print(first_img.shape)
            #print(last_img.shape)
            
            
            if len(first_img.shape) == 2:
                first_img = np.expand_dims(first_img, axis=-1)
            
            if len(last_img.shape) == 2:
                last_img = np.expand_dims(last_img, axis=-1)
            
            img_seq = []
            img_seq.append(first_img)
            img_seq.append(last_img)
            img_seq = np.array(img_seq)
            
            if img_seq.ndim == 3:
                img_seq = np.expand_dims(img_seq, axis=-1) 
            #print('New img-seq', img_seq.shape)
            xt = np.transpose(img_seq,(1,0,2,3))
            #print(xt.shape)
            # transpose the img_seq to extract feature from YT plane
            yt = np.transpose(img_seq,(2,0,1,3))
            # histogram of all frames in xy plane
            hist_xy_plane = np.empty(0,dtype=np.float32)
            # histogram of all frames in xt plane
            hist_xt_plane = np.empty(0,dtype=np.float32)
            # histogram of all frames in yt plane
            hist_yt_plane = np.empty(0,dtype=np.float32)
            
            n_frames_xy_plane = img_seq.shape[0]
            n_frames_xt_plane = xt.shape[0]
            n_frames_yt_plane = yt.shape[0]
            #print('this is xt shape', xt.shape)
            #print('this is yt shape', yt.shape)
        
        
                
            for i in range(n_frames_xy_plane):
           
                hist_xy, hog_xy = hog(img_seq[i], orientations=orientations, pixels_per_cell=pixels_per_cell,
                     cells_per_block=cells_per_block,block_norm = 'L2', visualize=True, channel_axis = -1) 
                hist_xy_plane = np.concatenate((hist_xy_plane,hist_xy))
                
            for i in range(n_frames_yt_plane):
                hist_yt,hog_yt = hog(yt[i], orientations=orientations, pixels_per_cell=(2,8),
                     cells_per_block=cells_per_block,block_norm ='L2', visualize=True,channel_axis = -1) #multichannel=True)
                hist_yt_plane = np.concatenate((hist_yt_plane,hist_yt))
            for i in range(n_frames_yt_plane):
                hist_yt,hog_yt = hog(yt[i], orientations=orientations, pixels_per_cell=(2,8),
                     cells_per_block=cells_per_block,block_norm ='L2', visualize=True,channel_axis = -1) #multichannel=True)
                hist_yt_plane = np.concatenate((hist_yt_plane,hist_yt))
            hog_top = np.concatenate((hist_xy_plane, hist_xt_plane, hist_yt_plane))
            hog_top = hog_top / np.linalg.norm(hog_top)
            
            hognum = hognum + 1
            print('Hog feature', hognum)
            #print(hog_top.dtype)   
            hogtop_features.append(hog_top)
        return hogtop_features
```
